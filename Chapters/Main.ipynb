#### Math Operator

```python
# Intersection
s1 = {0, 1, 2, 3}
s2 = {1, 3}

print(s1.intersection(s2))
print(s1 & s2)
```

```python
# Union 
s1 = {0, 1, 2, 3}
s2 = {1, 3, 4}

print(s1.union(s2))
print(s1 | s2)
```

```python
# Difference 
s1 = {0, 1, 2, 3}
s2 = {1, 3, 4}

print(s1.difference(s2))
print(s2.difference(s1))

print(s1 - s2, s2 - s1)
```

```python
# Symmetric Difference  
s1 = {0, 1, 2, 3}
s2 = {1, 3, 4}

print(s1.symmetric_difference(s2))

print(s1 ^ s2)
```

```python
# Product
s1 = {1, 2, 3}
s2 = {4, 5}

result = []
for x in s1:
    for y in s2:
        result.append((x, y))
        
print(result)


from itertools import product

print(list(product(s1, s2)))
```

#### Operator


```python
# Membership 
s1 = {1, 2, 3}

print(1 in s1)
print(2 in s1)
print(4 not in s1)
```

```python
# Disjoint   
s1 = {1, 2, 3}
s2 = {4, 5, 6, 2}

print(s1.intersection(s2) == set())
print(s1 & s2 == set())
print(s1.isdisjoint(s2))
```

```python
# Subset 
s1 = {4, 5, 6}
s2 = {4, 5, 6, 2}

print(s2.issuperset(s1))
print(s1.issubset(s2))
```

```python
# Superset  
s1 = {4, 5, 6}
s2 = {4, 5, 6, 2}

print(s2.issuperset(s1))
print(s1.issubset(s2))
```

```python
# Complement 
U = set(range(1, 11))
s1 = {4, 5, 6}

print(U - s1)
```

```python
# PowerSet  
from itertools import combinations as comb

s1 = {1, 2, 3, 4}
result = []

for lng in range(0, len(s1)+1):
    selection = list(comb(s1, lng))
    result.extend(selection)

for x in result:
    print(x)
```

```python
# Count each unique elements in a list            
nums = [1, 2, 1, 2, 3, 4, 4, 4, 3]

for x in set(nums):
    print(x, nums.count(x))

count = {} # dict()

for x in nums:
    if x in count:
        count[x] += 1
    else:
        count[x] = 1

print(count)
```
